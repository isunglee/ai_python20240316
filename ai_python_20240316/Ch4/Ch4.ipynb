# List VS Tuple

+ tuple 與 list 的差異
  + 括號不同：
    + tuple 是用小括號將元素包起來。
    + list 是用中括號將元素包起來。
+ 元素可否變更：
   + tuple 決定了元素的內容就不可以變更。
   + list 可以隨時修改

```python
myList = ['A',10,3.14]
myTuple = ('Ken',35)
print(type(myList))
print(type(myTuple))
```

```python
myList = ["Oen"]
print(myList)
myTuple = (10,)
print(myTuple)
print(type(myTuple))
```

```python
myList = ["A","B","C","D","E"]
myTuple = ("X","y",20,"H",15.623)
print(len(myList))
print(len(myTuple))
print(myList[:2])
print(myTuple[:3])
```

## 移除 
+ list 可透過以下三種方式移除項目：
  + list.remove( )
  + list.pop( )
  + del list[ ]

###  remove

依據內容進行刪除

list.remove(項目內容)

```python
list1 = ['1','x','2','x','3']
list1.remove('x')
print(list1)
list1.remove('x')
if 'x' in list1:
    list1.remove('x')
    
print(list1)
```

```python
list1 = ['1','x','2','x','3','y','c','g','h']
n = 'x'
while True:
    if n  in list1:
        list1.remove(n)
    else:
        break
print(list1)
```

### pop

如果 pop( ) 內沒有參數，則移除最後一筆資料。

pop() 內若有參數，則移除指定位置的資料。

```python
list2 = []
list2.append("Ken")
list2.append("Vivin")
list2.append("Lucy")
list2.append("Iris")
print(list2)
print(list2.pop())
print(list2.pop())
print(list2.pop())
print(list2.pop())

print(list2.pop())#list空了出錯
```

```python
list2 = []
list2.append("Ken")
list2.append("Vivin")
list2.append("Lucy")
list2.append("Iris")
while len(list2) > 0:
    print(list2.pop())
    
```

```python
list2 = []
list2.append("Ken")
list2.append("Vivin")
list2.append("Lucy")
list2.append("Iris")
while len(list2) > 0:
    print(list2.pop(0))
```

### Del
+ list 可透過 del 方式移除指定位置資料。
+ del list[ ] 必須指定範圍：
  + del list[n] 代表移除第 n 位索引值資料。
  + del list[:m] 代表移除第 m 位索引值之前 ( 不包含 m ) 資料。
  + del list[n:m] 代表移除第 n 到 m 位索引值之前 ( 不包含 m ) 資料。

```python
list2 = ["A","B","C","D","E"]
print(list2)
del list2[2]
print(list2)
del list2[1:3]
print(list2)
```
