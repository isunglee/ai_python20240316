```python
pm = int(input("輸入pm"))
if pm < 0:
    print("錯誤")
elif pm >=0 and pm <= 35:
    print("低")
elif pm >= 36 and  pm<= 53:
    print("中")
elif pm >= 54 and pm <= 70:
    print("高")
else:
    print("非常高")
#0~35 低
#36~53 中
#54~70 高
#>70非常高
```

## FOR迴圈
### 知道數量時使用for loop
### python的for 有一種由頭到取數值的想法


```python
r1 = range(10)
r1 = list(r1)
print(r1)
r2 = range(2,10)
r2 = list(r2)
print(r2)
r3 = range(1,20,2)
print(list(r3))
r4 = range(20,0,-2)
print(list(r4))
```

```python
for i in range(10): 
    print(i)
```

```python
for k in range(2,8):
    print(k,end=" ")
    #end每次呼叫print結尾輸出什麼
```

```python
for x in range(2,18,3):
    print(x,end = " ")
```

```python
for y in range(10,1,-2):
    print(y , end = " ")
```

```python
for i in range(4):
    print(i)
print("離開的i:",i)    
for x in range(i,-1,-1):
    print(x)
```

```python
a = 6
while a > 0:
    print(a)
    a-=1
```

```python
n = "y"
while n == "y":
    a = int(input("請輸入第一個數字"))    
    b = int(input("請輸入第二個數字"))    
    ans = a * b
    print(a,"*",b,"=",ans)
    n = input("是否繼續y/n:")
    n = n.lower()
```

## break 代表不回頭，跳出迴圈。
## continue 代表繼續執行下一個迴圈。

```python
for i in range(11):
    print(i)
    if i == 5:
        break
```

```python
for i in range(1,11):
    if i == 5 or i == 7:
        continue
    print(i,end=" ")
```
