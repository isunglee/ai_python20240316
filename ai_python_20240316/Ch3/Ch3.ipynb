```python
# BMI = 體重 / 身高^2 (單位公尺)
h = float(input("身高"))
w = int(input("體重"))
bmi = w / (h/100)**2
print(round(bmi,2) )
if bmi < 18.5:
    print("太輕")
elif bmi >= 18.5 and bmi <= 25:
    print("正常")
elif 26 <= bmi <= 30:
    print("過重")
else:
    print("胖胖")
```

## find

字串可以搜尋特定內容的位置。

搜尋的字串位置編號由 0 開始。

字串.find(搜尋字串,起始位置,結束位置)

搜尋時不包含結束位置。

若搜尋不到則回傳 -1。

若沒有起始位置，也沒有結束位置，預設由 0 開始。

若沒有結束位置，預設到最後一個字。

```python
str1 = "This is Python.That is Java,This is SQLite"
print(str1)
search1 = "Python"
print(str1.find(search1))
serach2 = "not"
print(str1.find(serach2))
serach3= "t"
print(str1.find(serach3))
serach4= "is"
print(str1.find(serach4,3))
```

```python
str1 = "This is Python.That is Java,This is SQLite"
search1 = "is"
index = 0
while True:
    index = str1.find(search1,index)
    if index == -1:
        break
    print("is:",index)
    index+=1
```

## join

join( ) : 於字串或著字元之間加入指定的文字

指定文字.join(要插入的字串)

```python
str1 = "-"
str2 = ("a","b","c")#tuple
print(str1.join(str2))
str3 = ("abc","xyz","123")
print(",".join(str3))
print(":".join("Python"))
```
