## 串接

```python
age = 15
##字串不能跟數字串接
##串接只能跟相同類型的
msg = "年齡:"+ age
print(msg)
```

```python
age = 15
#使用str將age轉換成文字
msg = "年齡:"+ str(age)
print(msg)
```

```python
a,b = 10,3
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)#無條件捨去
print(a % b)#取餘數(remainder)
ans = a / b
#round 取到小數點第n位
print(round(ans,2))
print(round(a/b,2))
```

```python
v1 = 2
ans = v1 ** 3  #2的3次方 second power
print(ans)

ans2 = v1 ** -1 #2的-1次方
print(ans2)

ans3 = v1 ** 0.5#2的1/2次方 開根號(Square root)
print(ans3)
```
